### Brand

```python
# unique values
print('Number of Unique Values: ', df.Brand.nunique()) 
print('Unique Values: ', df.Brand.unique())
```

```python
# target encode Brand with Screen size (inches) mean 
brand_screen_size = df.groupby('Brand')['Screen size (inches)'].sum().sort_values().to_dict() 
```

```python
df['Brand'] = df['Brand'].map(brand_screen_size)
```

```python
df.Brand.nunique()
```

### Categorical dummies Varibales 

```python
cat_features = ['Touchscreen', 'Operating system', 'Wi-Fi', 'Bluetooth', 'GPS', 'Number of SIMs', '3G', '4G/ LTE']
df = pd.get_dummies(df, columns = cat_features , drop_first=True) 
```

```python
df.head()
```

### Split the dataset

```python
# dependent and independent features
X = df.drop('Price', axis = 'columns') 
y = df['Price']
```

```python
#import the train_test_split 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
y_train.shape
```

### Standardize the Numerical Variables

```python
# import standardScaler 
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train

col_names= ['Brand', 'Battery capacity (mAh)', 'Screen size (inches)','Resolution x', 'Resolution y', 
                'Processor', 'RAM (MB)', 'Internal storage (GB)', 'Rear camera', 'Front camera'] 
```

```python
# standardize the X_train numerical variable
X_train[col_names] = scaler.fit_transform(X_train[col_names])
```

```python
# standardize the X_test numerical variable
X_test[col_names] = scaler.fit_transform(X_test[col_names])
```
